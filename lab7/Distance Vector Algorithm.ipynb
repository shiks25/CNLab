```python
class Graph:

    def __init__(self, vertices):
        self.V = vertices   
        self.graph = []   


    def add_edge(self, s, d, w):
        self.graph.append([s, d, w])


    def print_solution(self, dist, src, next_hop,res):
        print("Router Table for router ", res[src])
        print("Destination Router     Outgoing Line \t Hop Count")
        for i in range(self.V):
            print("\t{0} \t\t     {1} \t\t    {2}".format(res[i],res[next_hop[i]], dist[i]))

    def bellman_ford(self, src,res):

        dist = [99] * self.V
        dist[src] = 0
        next_hop = {src: src}
        for _ in range(self.V - 1):
            for s, d, w in self.graph:
                if dist[s] != 99 and dist[s] + w < dist[d]:
                    dist[d] = dist[s] + w
                    if s == src:
                        next_hop[d] =d
                    elif s in next_hop:
                        next_hop[d] = next_hop[s]
        
        for s, d, w in self.graph:
            if dist[s] != 99 and dist[s] + w < dist[d]:
                print("Graph contains negative weight cycle")
                return

        self.print_solution(dist, src, next_hop,res)
```

```python
def main():
    matrix = []
    print("Enter the no. of routers:")
    n = int(input())
    nodes = input('Enter the nodes: ').split()
    nums=list(range(0, n))
    res = {nums[i]: nodes[i] for i in range(len(nums))} 
    print("Enter the adjacency matrix : Enter 99 for infinity")
    for i in range(0,n):
        a = list(map(int, input().split(" ")))
        matrix.append(a)
        
    g = Graph(n)
    for i in range(0,n):
        for j in range(0,n):
            g.add_edge(i,j,matrix[i][j])
                
    for k in range(0, n):
        g.bellman_ford(k,res)
```
